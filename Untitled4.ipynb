```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("box_office (1).csv")
```

```python
plt.figure(figsize=(12, 8))
phase_grouped = df.groupby('Phase')['Budget'].mean().reset_index()
plt.plot(phase_grouped['Phase'], phase_grouped['Budget'], marker='o', color='green', linestyle='-')
plt.title('Distribuição do Tamanho do Orçamento por Fase do MCU')
plt.xlabel('Fase')
plt.ylabel('Orçamento de Produção Médio (em milhões)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Budget'], df['Box office gross Worldwide'], color='b', alpha=0.5)
plt.title('Correlação entre Orçamento e Bilheteria')
plt.xlabel('Orçamento de Produção (em milhões)')
plt.ylabel('Bilheteria Mundial (em bilhões)')
plt.grid(True)
plt.show()
```

```python
phase_grouped['Rotten Tomatoes Rating'].plot(kind='line', marker='o', figsize=(10, 6))
plt.title('Avaliação da Crítica por Fase')
plt.xlabel('Fase')
plt.ylabel('Avaliação Rotten Tomatoes')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
top_rated_movies = df.nlargest(10, 'Rotten Tomatoes Rating')

plt.figure(figsize=(12, 8))
plt.bar(top_rated_movies['Film'], top_rated_movies['Box office gross Worldwide'], color='blue')
plt.title('Bilheteria dos 10 Melhores Avaliados (Rotten Tomatoes)')
plt.xlabel('Filme')
plt.ylabel('Bilheteria Mundial (em bilhões)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
phase_grouped[['Box office gross U.S. and Canada', 'Box office gross Other territories']].plot(kind='bar', figsize=(10, 6))
plt.title('Bilheteria por Território por Fase')
plt.xlabel('Fase')
plt.ylabel('Bilheteria (em bilhões)')
plt.xticks(rotation=45)
plt.legend(['EUA/Canadá', 'Outros Territórios'])
plt.show()
```

```python
phase_grouped['Box office gross Worldwide'].plot(kind='bar', color='skyblue', figsize=(10, 6))
plt.title('Bilheteria Total por Fase')
plt.xlabel('Fase')
plt.ylabel('Bilheteria Mundial (em bilhões)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
director_grouped = df.groupby('Director').sum()
top_directors = director_grouped.sort_values(by='Box office gross Worldwide', ascending=False).head(10)
top_directors['Box office gross Worldwide'].plot(kind='bar', color='orange', figsize=(10, 6))
plt.title('Top 10 Diretores por Bilheteria')
plt.xlabel('Diretor')
plt.ylabel('Bilheteria Mundial (em bilhões)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
